```python
import time
import numpy as np
import pandas as pd
from  pandas.io.data import *
import hestonpricer as heston
import blackscholespricer as bs
from scipy.stats import norm
from scipy import optimize
```

```python
# input=market price, output=heston parameters by minimzimg the objective function
# output = [lambd, rho, eta, v0, v_bar]
def heston_calibration(optiondata):

    N = len(optiondata)
    
    S = optiondata['Spot']
    r = optiondata['iRate']
    q = optiondata['dRate']
    K = optiondata['Strike']
    market_vol = optiondata['IVMid']
    T = np.zeros(N)
    
    for i in range(0,N):
        ndays = (pd.to_datetime(optiondata['dtExpiry'].ix[i])-pd.to_datetime(optiondata['dtTrade'].ix[i])).days
        T[i] = ndays/360 
        
    market_price = np.zeros(N)
    
    for i in range(0,N):
        market_price[i] = bs.vanilla_price(S[i], K[i], r[i]-q[i], market_vol[i], T[i], 'C')
        
    model_price = np.zeros(N)
 
    # objective function is the root means squared error of vol
    sq_err = np.zeros(N) # sum of (market price - model price)**2
    def objfunc(params):
        lambd, rho, eta, v0, v_bar = params
        for i in range(0,N):
            model_price[i] = heston.callprice(S[i],K[i],r[i]-q[i],T[i],lambd,rho,eta,v0,v_bar)
            sq_err[i] = (market_price[i]-model_price[i])**2
        return sqrt((1/N)*sq_err.sum())
    
    # initial values of the parameters
    ival=(1,-0.5,0.5, 0.05, 0.01) # initial guess
    # bounds on each parameter
    bnd = ((0,10), (-1,1), (0,1), (0,1), (0,1))
    # using optimize.minimize to find optimal values of paramters
    #output = optimize.minimize(objfunc, ival, method='Nelder-Mead')
    output = optimize.minimize(objfunc, ival, method='SLSQP', bounds=bnd, options={'maxiter':10000})
    #output = optimize.minimize(objfunc, ival, method='L-BFGS-B', bounds=bnd)
    #output = optimize.leastsq(objfunc, ival)
    return output, model_price
    
    #return marketprice
    #return sqrt((1/N)*sq_err.sum())
```

```python
def market_price(optiondata):
    N = len(optiondata)
    S = optiondata['Spot']
    r = optiondata['iRate']
    q = optiondata['dRate']
    K = optiondata['Strike']
    market_vol = optiondata['IVMid']
    T = np.zeros(N)
    
    for i in range(0,N):
        ndays = (pd.to_datetime(optiondata['dtExpiry'].ix[i])-pd.to_datetime(optiondata['dtTrade'].ix[i])).days
        T[i]  = ndays/360
        
    market_price = np.zeros(N)
    for i in range(0,N):
        market_price[i] = bs.vanilla_price(S[i], K[i], r[i]-q[i], market_vol[i], T[i], 'C')
    return market_price
```
